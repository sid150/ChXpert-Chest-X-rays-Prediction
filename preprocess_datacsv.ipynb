```python
from tqdm import tqdm

# Enable tqdm for pandas apply
tqdm.pandas()

# Dataset base dirs (note: don't use quotes inside the strings)
base_dirs = [
    "/mnt/CheXpert-v1.0 batch 2 (train 1)",
    "/mnt/CheXpert-v1.0 batch 3 (train 2)"
]

# Extract relative path from 'Path'
df["relative_path"] = df["Path"].apply(lambda p: "/".join(p.split("/")[2:]))

# Function to find actual path or return NaN
def resolve_path(rel_path):
    for base in base_dirs:
        full_path = os.path.join(base, rel_path)
        if os.path.exists(full_path):
            return full_path
    return np.nan

# Apply with progress bar
df["corrected_path"] = df["relative_path"].progress_apply(resolve_path)

# Drop missing
filtered_df = df.dropna(subset=["corrected_path"]).drop(columns=["relative_path"])

# Save
filtered_df.to_csv("filtered_chexpert_paths.csv", index=False)
```

```python
import os

# Dataset batch root directories
base_dirs = [
    "/mnt/CheXpert-v1.0 batch 2 (train 1)",
    "/mnt/CheXpert-v1.0 batch 3 (train 2)",
    "/mnt/CheXpert-v1.0 batch 4 (train 3)"
]

# Total unique patient IDs across all batches
unique_patients = set()

# Dictionary to store per-batch patient counts
patient_counts = {}

for base in base_dirs:
    count = 0
    if os.path.exists(base):
        for entry in os.listdir(base):
            patient_path = os.path.join(base, entry)
            if os.path.isdir(patient_path) and entry.startswith("patient"):
                unique_patients.add(entry)
                count += 1
    patient_counts[os.path.basename(base)] = count

# Print per-directory counts
for batch, count in patient_counts.items():
    print(f"{batch}: {count} patients")

print(f"\nTotal unique patients across all batches: {len(unique_patients)}")
```

```python
import os

# Dataset batch root directories
base_dirs = [
    "/mnt/CheXpert-v1.0 batch 2 (train 1)",
    "/mnt/CheXpert-v1.0 batch 3 (train 2)",
    "/mnt/CheXpert-v1.0 batch 4 (train 3)"
]

jpg_count = 0

for base in base_dirs:
    if os.path.exists(base):
        for patient in os.listdir(base):
            patient_path = os.path.join(base, patient)
            if not os.path.isdir(patient_path) or not patient.startswith("patient"):
                continue
            for study in os.listdir(patient_path):
                study_path = os.path.join(patient_path, study)
                if not os.path.isdir(study_path):
                    continue
                for file in os.listdir(study_path):
                    if file.endswith(".jpg"):
                        jpg_count += 1

print(f"Total .jpg files found: {jpg_count}")
```

```python
import os

# Dataset batch root directories
base_dirs = [
    "/mnt/CheXpert-v1.0 batch 2 (train 1)",
    "/mnt/CheXpert-v1.0 batch 3 (train 2)",
    "/mnt/CheXpert-v1.0 batch 4 (train 3)"
]

jpg_count = 0
total_size_bytes = 0

for base in base_dirs:
    if os.path.exists(base):
        for patient in os.listdir(base):
            patient_path = os.path.join(base, patient)
            if not os.path.isdir(patient_path) or not patient.startswith("patient"):
                continue
            for study in os.listdir(patient_path):
                study_path = os.path.join(patient_path, study)
                if not os.path.isdir(study_path):
                    continue
                for file in os.listdir(study_path):
                    if file.endswith(".jpg"):
                        file_path = os.path.join(study_path, file)
                        if os.path.isfile(file_path):
                            jpg_count += 1
                            total_size_bytes += os.path.getsize(file_path)

# Convert to MB and GB
size_mb = total_size_bytes / (1024 ** 2)
size_gb = total_size_bytes / (1024 ** 3)

print(f"Total .jpg files found: {jpg_count}")
print(f"Total size: {total_size_bytes:,} bytes")
print(f"Total size: {size_mb:.2f} MB")
print(f"Total size: {size_gb:.2f} GB")
```
