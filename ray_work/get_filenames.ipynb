```python
import pandas as pd
import os
import numpy as np
```

```python
path = "/mnt/data/train_cheXbert.csv"
```

```python
!ls /mnt/data
```

```python
df = pd.read_csv(path)
```

```python
# Dataset base dirs
base_dirs = [
    "/mnt/data/CheXpert-v1.0 batch 2 (train 1)",
    "/mnt/data/CheXpert-v1.0 batch 3 (train 2)",
    "/mnt/data/CheXpert-v1.0 batch 4 (train 3)"
]

# Extract relative path from 'path'
df["relative_path"] = df["Path"].apply(lambda p: "/".join(p.split("/")[2:]))

# Function to find actual path or return NaN
def resolve_path(rel_path):
    for base in base_dirs:
        full_path = os.path.join(base, rel_path)
        if os.path.exists(full_path):
            return full_path
    return np.nan

# Vectorized mapping
df["corrected_path"] = df["relative_path"].apply(resolve_path)

# Drop missing files
filtered_df = df.dropna(subset=["corrected_path"]).drop(columns=["relative_path"])

# Save (optional)
filtered_df.to_csv("filtered_chexpert_paths.csv", index=False)
```
